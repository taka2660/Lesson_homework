```python
def calculator(a, b, operator):
    #関数の中身を考えてみてください(四則演算)
    if operator == '+':
      result = a + b
    elif operator == '-':
      result = a - b
    elif operator == '*':
      result = a * b
    elif operator == '/':
      result = a / b
    return result

a = float(input("1つ目の数値を入力してください: "))
b = float(input("2つ目の数値を入力してください: "))
operator = input("演算子を入力してください (+, -, *, /): ")
#四則演算する関数calculaorを呼び出してresultに代入
result = calculator(a, b, operator)
print("--------------------------------------------------")
print(f"{a} {operator} {b}の計算結果は{result}です。")


#関数を使う処理も記述しましょう。
#最終的に電卓の計算結果をprintで表示しましょう。
```
