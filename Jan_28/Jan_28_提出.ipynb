```python
"""
2種類のクラスの同名メソッドを使うプログラム
　・自動車シミュレーションプログラム
　→car_aというクラスを作成する。
　　→メンバ変数
　　　→速度
　　　→走行距離
　　　→走行時間
　　　→名前
　　→メソッド
　　　→速度を設定するメソッド(inputで入力しましょう)
　　　→走行時間を設定するメソッド(inputで入力しましょう)
           →(必殺技)例：他のキャラクターの速度を下げる
　→car_bというクラスを作成する。
　　→メンバ変数
　　　→速度
　　　→走行距離
　　　→走行時間
　　　→名前
　　→メソッド
　　　→速度を設定するメソッド(inputで入力しましょう)
　　　→走行時間を設定するメソッド(inputで入力しましょう)
           →(必殺技)例：自分の速度を上げる
　→car_aクラス、car_bから生成するオブジェクトの数は何台でもOK
　→一番進んでいる車の名前は？print文で出力
"""
#Car_Aクラスを定義
class car_A:
  def __init__(self, color):
    self.speed = 55.0
    self.distance = 0.0
    self.time = 0.0
    self.color = color

  def set_time(self):
    time_input = input(f"この車の速度は55.0km/hです。{self.color}の車の走行時間(時間)を入力してください (例: 1.5): ")
    self.time = float(time_input)

  def calc(self):
    self.distance = self.speed * self.time
    return self.distance

#Car_Bクラスを定義
class car_B:
  def __init__(self, color):
    self.speed = 63.0
    self.distance = 0.0
    self.time = 0.0
    self.color = color

  def set_time(self):
    time_input = input(f"この車の速度は63.0km/hです。{self.color}の車の走行時間(時間)を入力してください (例: 1.5): ")
    self.time = float(time_input)

  def calc(self):
    self.distance = self.speed * self.time
    return self.distance

#car_A, car_Bクラスから生成するオブジェクトのインスタンス化
car1 = car_A("黒色")
car2 = car_B("赤色")

## 生成したオブジェクトを配列carsのリストに格納
cars = [car1, car2]

#各車の走行距離を出力
# 各車の速度と時間を入力させ、距離を計算
for car in cars:  #リスト「cars」に入っている車を、1台ずつ取り出して car に入れる
    car.set_time()
    car.calc()
    print(f"{car.color}の車の走行距離は{car.distance}です。\n")

# 一番進んでいる車を判定
# max()関数とラムダ式を使って、distance属性が最大のオブジェクトを見つける
farthest_car = max(cars, key=lambda car: car.distance)

# 結果を出力
print(f"一番進んでいる車は{farthest_car.color}の車で走行距離は{farthest_car.distance}です。")
```
