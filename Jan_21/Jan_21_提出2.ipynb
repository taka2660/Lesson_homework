```python
'''
### 課題 1: 数当てゲーム
**目的**: ユーザーに対して1から100までのランダムな数字を当てさせるゲームを作成する。
**要件**:
- プログラムがランダムに1から100までの数字を生成。
- ユーザーに数字を入力させ、その入力に対して「大きい」か「小さい」かを返す。
- 正解したら「正解です！」と表示してゲーム終了。
'''
import random
def play_game():
  ran_select = random.randint(1, 100)  #ランダムに選定された1から100までの整数を取得する
  print("1から100までのランダムに選択された数字当ててください。")

#ランダム値と入力値の比較判定
  while True:
    #ユーザ値を入力する入力欄(number)を定義
    number = int(input("任意の1から100までの整数を入力して下さい："))
    if ran_select < number:
      print(f"残念( ；∀；)、入力した数字{number}はランダムに選択された数字より大きいです。")
    elif ran_select > number:
      print(f"残念( ；∀；)、入力した数字{number}はランダムに選択された数字より小さいです。")
    else: #入力値とランダム値が同じ場合
      print(f"入力した数字{number}はランダムに選択された数字{ran_select}と同じなので正解です！")
      break
play_game()
```
