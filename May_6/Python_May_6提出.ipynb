```python
import random

class SlotMachine:
    def __init__(self, initial_coins):
        #絵柄のリスト
        self.symbols = ['🍎', '🍋', '🍒', '💎']
        # 最初に持っているコイン数
        self.coins = initial_coins
        #今どの絵柄が並んでいるかを保存
        self.current_reels = []

    def spin(self, bet):
        #リールを回してコインを減らす
        #手持ちが賭け金（bet）より少ない場合は「回せない」判断
        if self.coins < bet:
            print("コインが足りません！")
            return False

        self.coins -= bet
        self.current_reels = [random.choice(self.symbols) for _ in range(3)]
        return True

    def check_result(self):
        #判定して配当を返す
        #つの絵柄をバラバラの変数に取り出し
        r1, r2, r3 = self.current_reels
        if r1 == r2 == r3:
            # 絵柄によって配当を変える 💎がそろったときは他の手と比べて高配当
            payout = 40 if r1 == '💎' else 10
            self.coins += payout
            return f"大当たり！ {payout}枚獲得！"
        return "ハズレ..."

    def show_status(self):
        #現在のリールと所持コインを表示
        reels_str = " | ".join(self.current_reels)
        print(f"[{reels_str}]  所持コイン: {self.coins}")

# ゲームの実行
machine = SlotMachine(initial_coins=100)

while machine.coins > 0:
    cmd = input("Enterでスピン (qでゲームを終了): ")
    if cmd == 'q': break

    if machine.spin(bet=1):
        machine.show_status()
        print(machine.check_result())
```
