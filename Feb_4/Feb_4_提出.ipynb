```python
class Car:
    def __init__(self, person, speed):
        self.speed = speed
        self.distance = 0.0
        self.time = 0.0
        self.person = person
        self.final_dest = 0.0

    def set_time(self):
        time_input = input(f"{self.person}の走行時間(時間)を入力してください (例: 1.5): ")
        self.time = float(time_input)

    def calc(self):
        self.distance = self.speed * self.time

    def dist_penalty(self, penalty):
        self.final_dest = self.distance - penalty


class CarA(Car):  #共通のクラス「Car」を継承して子クラスCarAを定義
    def __init__(self, person):
        super().__init__(person, 55.0) #走行速度は55kmで固定

    def dist_penalty(self):
        super().dist_penalty(20)
        return 20


class CarB(Car):  #共通のクラス「Car」を継承して子クラスCarBを定義
    def __init__(self, person):
        super().__init__(person, 63.0) #走行速度は63kmで固定

    def dist_penalty(self):
        super().dist_penalty(35)
        return 35


car1 = CarA("田中さんの車")
car2 = CarA("鈴木さんの車")
car3 = CarB("佐藤さんの車")
car4 = CarB("山田さんの車")

cars = [car1, car2, car3, car4]

for car in cars:
    car.set_time()
    car.calc()
    penalty = car.dist_penalty()

    print(f"{car.person}の走行距離は {car.distance} kmです。")
    print(f"必殺技の効果は -{penalty} kmです。")
    print(f"必殺技適用後の走行距離は {car.final_dest} kmです。\n")

farthest_car = max(cars, key=lambda car: car.final_dest)

print(f"一番進んでいる車は {farthest_car.person} で、必殺技適用後の走行距離は {farthest_car.final_dest} km です。")
```
