```python
"""
12/24提出課題
　→carというクラスを作成する。
　　→メンバ変数
　　　→速度
　　　→走行距離
　　　→走行時間
　　　→車体の色
　　→メソッド
　　　→速度を設定するメソッド(inputで入力しましょう)
　　　→走行時間を設定するメソッド(inputで入力しましょう)
　→carクラスから生成するオブジェクトの数は何台でもOK
　→一番進んでいる車は何色の車か？print文で出力
"""
#Carクラスを定義する：
class car:
  def __init__(self, color):
    #各メンバ変数を初期化
    self.speed = 0.0
    self.distance = 0.0
    self.time = 0.0
    self.color = color

  #クラスのメソッドの定義
  #メソッド1: 速度を設定するメソッド
  def set_speed(self):
    speed_input = input(f"{self.color}の車の走行速度(Km/h)を入力してください (例: 60.5): ")
    self.speed = float(speed_input)
  #メソッド2: 走行時間を設定するメソッド
  def set_time(self):
    time_input = input(f"{self.color}の車の走行時間(時間)を入力してください (例: 1.5): ")
    self.time = float(time_input)
  #メソッド3:入力された速度、時間から進んだ距離を計算するメソッド
  def calc(self):
    self.distance = self.speed * self.time
    return self.distance

#carクラスから生成するオブジェクトのインスタンス化
car1 = car("黒色")
car2 = car("赤色")
car3 = car("青色")

## 生成したオブジェクトを配列carsのリストに格納
cars = [car1, car2, car3]

#各車の走行距離を出力
# 各車の速度と時間を入力させ、距離を計算
for car in cars:
    car.set_speed()
    car.set_time()
    car.calc()
    print(f"{car.color}の車の走行距離は{car.distance}です。\n")

# 一番進んでいる車を判定
# max()関数とラムダ式を使って、distance属性が最大のオブジェクトを見つける
farthest_car = max(cars, key=lambda car: car.distance)

# 結果を出力
print(f"一番進んでいる車は{farthest_car.color}の車で走行距離は{car.distance}です。")


```
