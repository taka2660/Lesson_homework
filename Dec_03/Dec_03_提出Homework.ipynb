```python
"""
サイコロプログラムをオブジェクト化する。
　→diceというクラスをつくる。
　　→サイコロを振る関数
　　→サイコロの色情報を持つ変数
　　→合計値を保存する変数
　　→15回振った結果を入れるリスト
　→色は赤、青、黄色
　→最後に、色ごとの合計値を表示する(可能なら判定も)。
"""

import random

class dice:
    def __init__(self, color):
        self.color = color  # サイコロの色の属性
        self.results = []  # 15回振った結果を格納するリスト
        self.total = 0  # 合計値を保存する変数

    # サイコロを振る関数
    def roll(self):
        result = random.randint(1, 6)
        self.results.append(result)
        self.total += result
        return result

# 実行回数を定義
rolling = 15

# 色ごとのサイコロのインスタンスを定義
red_dice = dice("赤")
blue_dice = dice("青")
yellow_dice = dice("黄")


for i in range(rolling):
    # 各サイコロのrollメソッドを呼び出し、出目を一時変数に格納
    red_result = red_dice.roll()
    blue_result = blue_dice.roll()
    yellow_result = yellow_dice.roll()
    #各色のサイコロを15回振った合計を算出
    sum_red = 1 + red_result
    sum_blue = 1 + blue_result
    sum_yellow = 1 + yellow_result
    #各色サイコロを振った毎の結果を色ごとに表示
    print(f"{i+1}回目の赤色サイコロを振った結果は{red_result}でした：{i+1}回目の青色サイコロを振った結果は{blue_result}でした：{i+1}回目の黄色サイコロを振った結果は{yellow_result}でした")
#各色のサイコロを15回振った合計を最後の行に表示
#print(f"赤色サイコロを15回振った合計は{red_dice.total}でした。　青色サイコロを15回振った合計は{blue_dice.total}でした。　黄色サイコロを15回振った合計は{yellow_dice.total}でした。")

#各色のサイコロを15回振った合計と75点以上は高得点、75点以下は通常得点と判断して表記する
#赤色
score_red = 75
if red_dice.total >= 75:
  print(f"赤色のサイコロ15回を振った合計は{red_dice.total}だったので高得点です。")
else:
  print(f"赤色のサイコロ15回を振った合計は{red_dice.total}だったので通常得点です。")
#青色
if blue_dice.total >= 75:
  print(f"青色のサイコロを15回振った合計は{blue_dice.total}だったので高得点です。")
else:
  print(f"青色のサイコロを15回振った合計は{blue_dice.total}だったので通常得点です。")
#黄色
if yellow_dice.total >= 75:
  print(f"青色のサイコロを15回振った合計は{yellow_dice.total}だったので高得点です。")
else:
  print(f"黄色のサイコロを15回振った合計は{yellow_dice.total}だったので通常得点です。")

```
