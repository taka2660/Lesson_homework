```python
"""
　→Productというクラスを作成する。
　　→メンバ変数
　　　→商品名
　　　→価格
　　　→カテゴリ★今回は使わない。
　　　→在庫数
　　→メソッド
　　　在庫数を更新する関数
　→製品、カメラ、4品をオブジェクトで生成
　→とある日の売り上げ管理。売り上げた数をランダムに生成。
　→売上高が一番高かった製品をprintで出力する。
　　(例：一番売上高が高かったのは○○で、■■円でした。)
"""

import random

class Product:
    #Productクラスの詳細を定義：name, price, stock
    def __init__(self, name: str, price: int, stock: int):
        self.name = name
        self.price = price
        self.stock = stock

    def update_stock(self, quantity_sold: int):
        #在庫数を更新するメソッド、在庫数が不足する場合はエラーを返す
        if self.stock >= quantity_sold:
            self.stock -= quantity_sold
            return True
        else:
            print(f"エラー: 在庫が不足しています。")

# インスタンス化してオブジェクトを生成
# 製品名, 価格, 初期在庫数
product_N = Product("ニコン", 300000, 40)
product_C = Product("キャノン", 350000, 60)
product_P = Product("ペンタックス", 250000, 25)
product_S = Product("ソニー", 325000, 35)

#productオブジェクトを格納するリストを定義
products = [product_N, product_C, product_P, product_S]

# 売上データを格納するディレクトリを定義
sales_data = {}

print("--- とある日の売上管理開始 ---")

# 各製品の売り上げた数をランダムに生成し、在庫を更新
for product in products:
    # 売上げ数を0から10までの値でランダムに生成
    quantity_sold = random.randint(0, 10)

    if product.update_stock(quantity_sold):
        # 売上高を計算して記録
        daily_sales = quantity_sold * product.price
        sales_data[product.name] = daily_sales
        print(f"{product.name}: {quantity_sold}個販売, 売上高: {daily_sales}円, 新在庫数: {product.stock}")

# 売上高が一番高かった製品を特定し、出力する
if sales_data:
    # sales_dataディクショナリで定義した値を基に、最も売上高が高いキー（製品名）を取得
    best_seller_name = max(sales_data, key=sales_data.get)
    highest_sales = sales_data[best_seller_name]

    print("--- 結果 ---")
    print(f"売上高が一番高かった製品: {best_seller_name}")
    print(f"その売上高: {highest_sales}円")
else:
    print("売上データがありませんでした。")

```
