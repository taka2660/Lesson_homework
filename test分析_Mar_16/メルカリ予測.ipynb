```python
# ライブラリ
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder  #文字列を 数値に変換
from sklearn.ensemble import RandomForestRegressor  #回帰モデル（価格予測)
```

```python
# 1 データ読み込み
train = pd.read_csv("train.tsv", sep="\t")
test = pd.read_csv("test.tsv", sep="\t")
```

```python
# 2. データ確認
train.head()
train.info()
train.describe()
```

```python
# 3. 欠損確認
train.isnull().sum()
```

```python
# 4. 欠損箇所処理
for col in ["category_name","brand_name","item_description"]:
    train[col] = train[col].fillna("None")
    test[col] = test[col].fillna("None")
```

```python
# 5. 再確認
train.isnull().sum()
```

```python
# 6 カテゴリ変数の数値化
le_brand = LabelEncoder()
le_cat = LabelEncoder()

le_brand.fit(pd.concat([train["brand_name"], test["brand_name"]]))
train["brand_name"] = le_brand.transform(train["brand_name"])
test["brand_name"] = le_brand.transform(test["brand_name"])

le_cat.fit(pd.concat([train["category_name"], test["category_name"]]))
train["category_name"] = le_cat.transform(train["category_name"])
test["category_name"] = le_cat.transform(test["category_name"])
```

```python
# 7. 説明変数作成
drop_cols = ["train_id","name","item_description","price"]

X = train.drop(drop_cols, axis=1)
y = train["price"]

test_X = test.drop(["test_id","name","item_description"], axis=1)
```

```python
# 8. モデル作成
model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

model.fit(X,y)
```

```python
# 6 予測
pred = model.predict(test_X)
```

```python
# 7 提出ファイル
submission = pd.DataFrame({
    "test_id":test["test_id"],
    "price":pred
})

submission.to_csv("submission2.csv",index=False)

print("submission.csv を保存しました")
```
