```python
import random

class Characters:
    def __init__(self, name, hp=200, attck=10):
        self.name = name
        self.hp = hp
        self.attck = attck

    #Characterの生存判定hpが0以上であるか否かを判定
    def life_judge(self):
        return self.hp > 0

    #攻撃を受けたときのhp処理
    def take_damage(self, damage):
        self.hp -= damage
        if self.hp < 0:
            self.hp = 0

    #攻撃を受けた内容を表示
    def attck_enemy(self, enemy):
        enemy.take_damage(self.attck)
        print(f"{self.name}の攻撃！{enemy.name}に{self.attck}ダメージ！")

#Plyerのhp回復条件を親クラスを継承
class Player(Characters):
    def life(self):
        selection = random.randint(1, 3)  #攻撃力upはrandomで決定
        if selection == 1:
            life_hp = 5
        elif selection == 2:
            life_hp = 0
        else:
            life_hp = 15

        self.hp += life_hp
        print(f"{self.name}はHPを{life_hp}回復した！")

#Monsetrの攻撃力up条件を親クラスを継承
class Monster(Characters):
    def special_attack(self, enemy):
        selection = random.randint(1, 3)  #hp回復はrandomで決定
        if selection == 1:
            power = self.attck
        elif selection == 2:
            power = int(self.attck * 1.5)
        else:
            power = int(self.attck * 2)

        enemy.take_damage(power)
        print(f"{self.name}の攻撃！{enemy.name}に{power}ダメージ！")


player = Player("プレーヤー")
monster = Monster("モンスター")

round_num = 1

while player.life_judge() and monster.life_judge():
  print(f"\n{round_num}回目のラウンド！")

# プレイヤーの攻撃
  player.attck_enemy(monster)
  print(f"攻撃を受けたモンスターのHPは{monster.hp}です。")


# モンスターの攻撃
  monster.special_attack(player)
  print(f"攻撃を受けたプレーヤーのHPは{player.hp}です。")

  player.life()

  if not player.life_judge():
      print("プレーヤーは死にました、ゲームオーバー…")
      break
  round_num += 1
```
